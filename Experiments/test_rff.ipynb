```python
from Experiments.RFF import RFFSampler
import numpy as np
import torch
import matplotlib.pyplot as plt
```

```python
sampler = RFFSampler(num_features=1000, 
                     input_dim=1, 
                     number_of_functions=2, 
                     lengthscale=1.0, 
                     kernel="Matern52")

decision_variables = torch.linspace(-5, 5, 10000, dtype=torch.float64).reshape(-1, 1)  # Generate decision space samples

function_samples = sampler.sample(decision_variables)
function_samples.shape
```

```python
plt.plot(decision_variables.numpy(), function_samples[:, 0].numpy(), label='Function 1')
plt.plot(decision_variables.numpy(), function_samples[:, 1].numpy(), label='Function 2')
plt.xlabel('Decision Variables')
plt.ylabel('Function Values')
plt.title('RFF Samples')
plt.legend()
plt.show()
```
